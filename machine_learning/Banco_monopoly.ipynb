```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

col_cat = ['Subsegmento', 'Sexo', 'Region', 'Edad', 'Internauta', 'Adicional']

# Graficos de barras
fig, axs = plt.subplots(nrows=len(col_cat), ncols=1, figsize=(10, 45))
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(col_cat):
    sns.countplot(data=df, x=col, ax=axs[i])
    axs[i].set_title(f'Conteo de valores para {col}')
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('Conteo')

plt.show()
```

```python
# Y deseas crear un histograma para la columna Renta
columnas = ['Renta']

# Configura el tamaño de la figura
plt.figure(figsize=(12, 5))

# Itera a través de las columnas y crea un histograma para cada una
for columna in columnas:
    plt.hist(df[columna], bins=20, alpha=0.5, label=columna)

# Personaliza el gráfico
plt.title('Histograma de Renta')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()

# Muestra el gráfico
plt.show()
```

```python
# Y deseas crear un histograma para la columna Renta
columnas = ['Antiguedad']

# Configura el tamaño de la figura
plt.figure(figsize=(12, 5))

# Itera a través de las columnas y crea un histograma para cada una
for columna in columnas:
    plt.hist(df[columna], bins=20, alpha=0.5, label=columna)

# Personaliza el gráfico
plt.title('Histograma de Antiguedad')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()

# Muestra el gráfico
plt.show()
```

```python
#desviacion estandar de renta

renta_data = df['Renta']

media = renta_data.mean()
desviacion_estandar = renta_data.std()

x = np.linspace(renta_data.min(), renta_data.max(), 1000)
pdf = (1 / (desviacion_estandar * np.sqrt(2 * np.pi))) * np.exp(-((x - media) ** 2) / (2 * desviacion_estandar ** 2))

plt.figure(figsize=(10, 5))
plt.plot(x, pdf, label='Campana de Gauss', color='blue')
plt.xlabel('Renta')
plt.ylabel('Densidad de Probabilidad')
plt.title('Distribución de Renta con Campana de Gauss')
plt.legend()
plt.grid(True)
plt.show()

print("La moda de la columna 'Renta' es:")
print(df['Renta'].mode())
```
